```python
# import everything we need
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# read in the dataset
df = pd.read_csv('WASI_JND_DATASET_N106.csv')

# drop rows that don't have both jnd and matrix reasoning
df_clean = df[['JND_75', 'MatrixReasoning']].dropna()

# set a clean plot style
sns.set(style="whitegrid")

# create the scatterplot with a custom color palette (because i want it to look pretty)
plt.figure(figsize=(8, 6))
sns.regplot(
    x='MatrixReasoning',
    y='JND_75',
    data=df_clean,
    ci=95,
    scatter_kws={'s': 60, 'color': '#6BAED6'},  # light blue points
    line_kws={'color': '#08519C', 'linewidth': 2}  # darker blue line
)
plt.title('JND Threshold vs Matrix Reasoning Score', fontsize=14)
plt.xlabel('Matrix Reasoning Score', fontsize=12)
plt.ylabel('JND Threshold (°)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# run the correlation and print it
r, p = pearsonr(df_clean['MatrixReasoning'], df_clean['JND_75'])
print(f"Pearson correlation (r): {r:.3f}")
print(f"p-value: {p:.4f}")
```
